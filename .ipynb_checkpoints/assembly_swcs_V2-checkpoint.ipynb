```python
import os
from netpyne import specs, sim
%matplotlib
```

```python
DIR = 'assembly_swcs_V2'

swc_filenames = []
for filename in os.listdir(DIR):
    if os.path.splitext(filename)[-1] == '.swc':
        swc_filenames.append(os.path.join(DIR, filename))
        
assert len(swc_filenames)>0, 'No SWC files found!'
        
netParams = specs.NetParams()

for ii, swc_filename in enumerate(swc_filenames):
    cellRule = netParams.importCellParams(
        label='swc_cell{}'.format(ii), 
        fileName=swc_filename, 
        cellName='BS0284',
        )
```
